```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# import data
data = pd.read_csv('/content/UII.csv')
```

```python
# Show the first five entries of the dataset
print(data.head())
```

```python
# Show us the information regarding the selected dataset
data.info
```

```python
# Rename all the given columns
data.columns=["States", "Date", "Frequency", "Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate", "Region"]
```

```python
# Look at the correlation between the features of this dataset.
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(15,12))
sns.heatmap(data.corr())
plt.show()
```

```python
# Look at the estimated number of employees according to different regions of India
data.columns=["States", "Date", "Frequency", "Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate", "Region"]
plt.title("Indian Unemployment Scenario")
sns.histplot(x="Estimated Employed",hue="Region",data=data)
plt.show()
```

```python
# See at the unemployment rate according to the different states of India
plt.figure(figsize=(18,15))
plt.title("Indian Unemployment Scenario")
sns.histplot(x="Estimated Unemployment Rate",hue="States",data=data)
plt.show()
```
